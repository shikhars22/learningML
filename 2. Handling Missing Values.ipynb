# Filling or assigning missing values in a dataset = Imputation

## Wrong way of filling data to handle missing values in categorical data

```python
#dataset.fillna(10).head(10)
```

## Correct ways of filling data to handle missing values in categorical data or time-series data

### Identify categorical data

```python
dataset.info()
```

### Backfill data

```python
dataset["Gender"] = dataset["Gender"].bfill()
dataset["Gender"].isnull().sum()
```

### Forward Fill data

```python
dataset["Subscription"] = dataset["Subscription"].ffill()
dataset["Subscription"].isnull().sum()
```

### Forward Fill data with axis - example

```python
dataset[["DeptCode", "DiscountCode"]].ffill(axis=1)
```

### Mode Imputation

```python
dataset["DiscountCode"].mode()[0]
```

```python
dataset["DiscountCode"] = dataset["DiscountCode"].fillna(dataset["DiscountCode"].mode()[0])
dataset["DiscountCode"]
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer(strategy="mean")
ar = si.fit_transform(dataset[['Age', 'Fare', 'Score']])
ar
```

```python
new_dataset = pd.DataFrame(ar,columns = dataset.select_dtypes(include="float64").columns)
new_dataset
```

```python
new_dataset.isnull().sum()
```

```python
dataset['Age'].isnull().sum()
```

#### Insert the valules into the original dataset

```python
cols_to_fill = ['Age', 'Fare', 'Score']
dataset[cols_to_fill] = ar
print(dataset[cols_to_fill].isnull().sum())
```
